# 6.where do our majority customers fall(age)?

```python
x=df['Age']
plt.hist(x,bins=25, color= ['red'], edgecolor = 'black', linestyle = '--', alpha=0.60)
plt.title('Majority Customers')
plt.ylabel('Frequency')
plt.xlabel('Age')
plt.show()
```

# 7.Which state has highest sales?

```python

state_sales=df.groupby('State')['Amount'].sum()
state_sales
sns.countplot(x='State', data=df)

plt.title('Number of Users by State')
plt.xlabel('State')
plt.ylabel('Number of Users')
```

# 8 . How does occupation effect the purchase?

```python

occupation_sales = df.groupby('Occupation')['Amount'].sum()
occupation_sales
sns.countplot(x='Occupation', data=df)

plt.title('Number of Users by occupation')
plt.xlabel('occupation')
plt.ylabel('Amount')
```
